```python
import numpy as np
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree

df=pd.read_csv('/content/drive/MyDrive/dataset_decision_tree.csv')
```

```python
df=df.drop('sum',axis=1)
df
```

```python
###Extract features and labels

X=df.drop('Result',axis=1)
y=df['Result']
X.shape
```

```python
####Feature Scaling
scaler=StandardScaler()
scaler.fit(X)
strnd_data=scaler.fit_transform(X)
X=strnd_data
y=df['Result']
```

```python
###spliting the data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf_entropy=DecisionTreeClassifier(criterion='entropy',random_state=42,max_depth=5,min_samples_leaf=8)
clf_entropy.fit(X_train,y_train)
```

```python
y_pred=clf_entropy.predict(X_test)
y_pred
```

```python
print('Accuracy of model is:',accuracy_score(y_test,y_pred)*100)
```

```python
tree.plot_tree(clf_entropy.fit(X_test, y_test))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(30, 10))
plot_tree(
    clf_entropy,
    feature_names=['Initial payment', 'Last payment', 'Credit score', 'House Number'],
    impurity=False,
    filled=True,
    fontsize=17
)
plt.show()
```
